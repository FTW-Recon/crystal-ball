```python
def home_winner(score):
    score_list = score.split()
    home = int(score_list[0])
    away = int(score_list[2])
    
    if home > away:
        return True
    return False

matches["home_win"] = matches["score"].apply(lambda s: home_winner(s))
matches
```

```python
h_team = matches[["home_team", "round"]].rename(columns={"home_team": "id_team"})
a_team = matches[["away_team", "round"]].rename(columns={"away_team": "id_team"})

team = pd.concat([h_team, a_team])
team
```

```python
def get_opponent(row):
    id_team = row["id_team"]
    m_round = row["round"]
    
    all_teams = matches.loc[
        (matches["round"] == m_round)
        & (
            (matches.home_team == id_team)
            | (matches.away_team == id_team)
        ),
        ["home_team", "away_team"]
    ].values[0]

    if len(all_teams) != 2:
        raise Exception("many values for opponent")
        
    opponent = all_teams[0] if all_teams[1] == id_team else all_teams[0]
    
    return opponent


team["opponent"] = team.apply(lambda row: get_opponent(row), axis=1)
```

```python
def is_home_team(row):
    id_team = row["id_team"]
    m_round = row["round"]
    
    home_team = matches.loc[
        (matches["round"] == m_round)
        & (matches.home_team == id_team)
    ].shape[0] == 1
    
    return home_team


team["was_home_team"] = team.apply(lambda row: is_home_team(row), axis=1)
```

```python
def is_winner(row):
    id_team = row["id_team"]
    m_round = row["round"]
    
    home_team = row["was_home_team"]
    location = "home_team" if home_team else "away_team"
    
    winner = matches.loc[
        (matches["round"] == m_round)
        & (matches[location] == id_team),
        "home_win"
    ].values[0]
    
    return winner

team["has_won"] = team.apply(lambda row: is_winner(row), axis=1)
```

```python
def get_date(row):
    id_team = row["id_team"]
    m_round = row["round"]
    
    date = matches.loc[
        (matches["round"] == m_round)
        & (
            (matches.home_team == id_team)
            | (matches.away_team == id_team)
        ),
        "date"
    ].values[0]
    
    return date

team["date"] = team.apply(lambda row: get_date(row), axis=1)
```
